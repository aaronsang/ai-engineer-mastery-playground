```python
data.isnull() # returns boolean same size object indicating if null values present 
```

```python
data.isnull().any() # chaining with .any() identifies if a column contains null values or not
```

```python
data.isnull().sum() # chaining with .sum() counts how many null values exist within a column
```

```python
data.isnull().sum() / data.shape[0] # displays null values per column in a percentage
```

```python
data.loc[:,'type'] # view type column
```

```python
data.loc[:,'type'].unique() # displays each unique value in the type column
```

```python
data.loc[:,'type'].value_counts(dropna=False) # display sums of all values in the type column
```

```python
data.loc[:,'type'].value_counts() / data.loc[:,'type'].notnull().sum() # returns each value in percent of non-null rows
```

```python
data.loc[:,'type'].notnull().sum() # number of rows where the type label/column is not null
```
