# Unemployment Rate Analysis - Learning Matplotlib 

The goal of this project is to learn Matplotlib.

```python
import pandas as pd
# how to import the pyplot module from matplotlib
import matplotlib.pyplot as plt
```

```python
unrate = pd.read_csv('https://raw.githubusercontent.com/aaronsang/guided-data-science-projects/master/data/unrate.csv')
```

```python
# lower case of column labels
unrate.columns = unrate.columns.str.lower()
# convert date column to datetime object
unrate['date'] = pd.to_datetime(unrate['date'])
```

```python
# 
plt.plot(unrate['date'].iloc[:12], unrate['value'].iloc[:12])
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
```

## Figures, subplots

```python
fig = plt.figure() # Creates figure which contains the axes
ax1 = fig.add_subplot(2,1,1) # axes object
ax2 = fig.add_subplot(2,1,2) # axes object
ax1.plot(unrate['date'].iloc[:12], unrate['value'].iloc[:12])
ax2.plot(unrate['date'].iloc[12:24], unrate['value'].iloc[12:24])
plt.show()
```
